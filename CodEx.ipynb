```python
VOYELLES = ["a", "e", "i", "o", "u", "y"]

def dentiste(texte):
    dent= ''
    for lettre in texte :
        if lettre in VOYELLES :
            dent+=lettre
    return dent
```

```python
dentiste("j'ai mal")
```
