Quelles étaient les raisons de cet hommage rendu par Google ?

>  Car les travaux de George Boole ont joué un rôle important dans la révolution numérique du 20e siècle. Considéré comme le créateur de la logique moderne, le mathématicien britannique a créé une structure algébrique et sémantique, utilisée encore aujourd'hui dans la programmation.

Qu'est-ce qu'une variable booléenne ?

> Une variable booléenne est une quantité logique qui n'admet que deux états possibles : « Vrai » ou « Faux »

Qu'elles sont les principales fonctions (opérations) booléennes ?

> and / or / not 

Expliquer en observant le gif animé du doodle ci-dessus, les conditions pour que :

- le G s'allume en bleu  :

> x et y == True

- le deuxième o s'allume en jaune :

> x ou y == True

- le l s'allume en vert :

> y == False

- le e s'allume en rouge :

> x == False

- le premier o s'allume en rouge :

> x == True ou y == False mais pas les deux en meme temps
