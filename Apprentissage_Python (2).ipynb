# if et else :

```python
if True:
    print('Ceci sera affiché')

if False:
    print('Ceci non')
```

En général, une structure conditionnelle if ressemble à ça :  

if 'condition':  
        'corps'


Où condition est une expression qui s'évalue en un booléen et corps est une suite d'une ou plusieurs instructions indentées. La structure est assez similaire à une boucle for. Remarquez les deux points (:) qui sont essentiels.

Quand l'ordinateur voit if condition:, il vérifie si condition est vraie (True). Si c'est le cas, il exécute le corps : un bloc d'instructions. Sinon, il l'évite et passe à la suite du script.

```python
phrase = 'Hello World'
ravi = True
if ravi:
    phrase += '!'
print(phrase)
```

```python
phrase = 'Hello World'
ravi = False
if ravi:
    phrase += '!'
print(phrase)
```

```python
phrase = 'Hello World'
ravi = True 
confus = True

if confus :
    phrase += '?'
if confus :
    if ravi :
        phrase += '!'
   
print(phrase)
```

# Changement caractère :

```python
phrase = 'HELLO WORLD'
phrase = phrase.lower()
phrase = phrase.capitalize()
print(phrase)
```

```python
phrase = 'un dernier exercice et vous pouvez vous reposer.'
new = ''
maj = True

for caractere in phrase :
    
    if maj :
        caractere = caractere.upper()
        new+=caractere
        maj = False
    else : 
        caractere=caractere.lower()
        new+=caractere
        maj = True
        
print(new)
```

```python
nom = 'Melissa'
nouveau_nom = ''
for c in nom:
    if c == 's':
        c = '$'
    
    
    if c == 'e':
        c = '3'
   
    
    if c == 'a':
        c = '@'

    nouveau_nom += c



print(nouveau_nom)
```

# Elif :

```python
adn = 'AGTAGCGTC'
adn_inverse = ''
for caractere in adn:
    if caractere == 'A':
        caractere = 'T'
    elif caractere == 'T':
        caractere = 'A'
    elif caractere == 'G':
        caractere = 'C'
    elif caractere == 'C':
        caractere = 'G'
    adn_inverse += caractere

print(adn_inverse)
```

```python
pourcentage = 73

if pourcentage < 40:
    note = 'F'
elif pourcentage < 60:
    note = 'C'
elif pourcentage < 80:
    note = 'B'
else:
    note = 'A'

print(note)
```
